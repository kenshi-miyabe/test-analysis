# 使い方

10問の中間試験を行い，隣の人と交換するなどして，採点してもらう．

Oh-o!Meijiのクラスウェブのアンケートで，

- 単一選択形式(ラジオボタン)
- 回答必須
- 正解10点，不正解0点

の設問を10個作って，各自登録してもらう．

結果ダウンロードしたExcelファイルをアップロードし，以下のスクリプトを実行する．

必要に応じて，ファイル名，集計する列のリスト，読み込むページを変更する

注意：名前や番号の部分を消してからアップロードするなど，プライバシーには配慮しましょう．

```python
# 読み込むExcelファイル名
file_name = "mid-test-A.xlsx"

# 集計する列のリスト，11行目から公差3で長さ10の数列の場合は以下
cols = [10 + i * 3 for i in range(10)]
```

```python
import pandas as pd
# ファイルの2ページ目からデータを読み込む
df = pd.read_excel(file_name, sheet_name=1, usecols=cols)
# NaNが存在する行は削除
df = df.dropna()
# 各行の合計を計算
df['total'] = df.sum(axis=1)
df
```

```python
# 主要な統計量を計算
mean_total = df['total'].mean()
std_total = df['total'].std()
min_total = df['total'].min()
max_total = df['total'].max()
median_total = df['total'].median()
```

```python
# 受験者数
num_rows = len(df)
# 各問題の平均
column_means = df.mean()

# 結果を整形
result = f"受験者数: {len(df)}名\n"
result += f"合計点の平均: {mean_total:.2f}点\n"
result += f"合計点の標準偏差: {std_total:.2f}点\n"
result += f"合計点の最小値: {min_total:.2f}点\n"
result += f"合計点の最大値: {max_total:.2f}点\n"
result += f"合計点の中央値: {median_total:.2f}点\n"

for i in range(10):
    problem_mean = column_means[i]
    result += f"問題{i+1}の平均点: {problem_mean:.2f}点\n"

# txtファイルとして保存
with open('result.txt', 'w') as file:
    file.write(result)

print("結果がresult.txtファイルに保存されました。")
```
